**Q1: Tokenization and Filtering (Stop Words & Punctuation)**

**Step 1.1: Load the Data_1.txt Text Content**

The text corpus from `Data_1.txt` is embedded directly below into a string variable.


```python
corpus_text = """Classification is the task of choosing the correct class label for a given input. In basic
classification tasks, each input is considered in isolation from all other inputs, and the set of labels is defined in advance. The basic classification task has a number of interesting variants. For example, in multiclass classification, each instance may be assigned multiple labels; in open-class classification, the set of labels is not defined in advance; and in sequence classification, a list of inputs are jointly classified."""

print("Original Text Content:\n")
print(corpus_text)
```

**Step 1.2: Demonstrate Three Separate Tokenization Methods**

We perform tokenization using basic Python string splits, regular expressions (`re`), and the standard NLTK tokenization engine to extract individual elements.

```python
import re
from nltk.tokenize import word_tokenize

tokens_split = corpus_text.split()
print(f"1. Built-in .split() [Count: {len(tokens_split)} tokens]:\n", tokens_split, "\n")

tokens_regex = re.findall(r'\b\w+\b', corpus_text)
print(f"2. Regular Expression [Count: {len(tokens_regex)} tokens]:\n", tokens_regex, "\n")

tokens_nltk = word_tokenize(corpus_text)
print(f"3. NLTK word_tokenize [Count: {len(tokens_nltk)} tokens]:\n", tokens_nltk)
```

**Step 1.3: Track and Remove Stop Words & Punctuation**

Using NLTK's English stop word repository and standard punctuation mappings, we capture, identify, and clear non-essential filler details from the NLTK tokens.

```python
import string
from nltk.corpus import stopwords

stop_words_list = set(stopwords.words('english'))
punctuation_list = set(string.punctuation)

found_stopwords = []
found_punctuations = []
cleaned_tokens = []

for word in tokens_nltk:
    word_lower = word.lower()
    if word_lower in stop_words_list:
        found_stopwords.append(word)
    elif word in punctuation_list:
        found_punctuations.append(word)
    else:
        cleaned_tokens.append(word)

print("Distinct Stop Words Found:\n", sorted(list(set(found_stopwords))), "\n")
print("Distinct Punctuations Found:\n", sorted(list(set(found_punctuations))), "\n")
print("Cleaned Tokens List (Filtered Output):\n", cleaned_tokens)
```

**Q2: Word Stemming**

**Step 2.1: Compare Regular Expression, Porter, and Lancaster Stemmers**

We subject our cleaned token list to three distinct structural stemming techniques to evaluate how they handle suffix truncation and token reduction.

```python
from nltk.stem import RegexpStemmer, PorterStemmer, LancasterStemmer

regex_stemmer = RegexpStemmer('ing$|ed$|s$', min=4)
porter_stemmer = PorterStemmer()
lancaster_stemmer = LancasterStemmer()

print(f"{'Original Word':<20} | {'RegEx Stemmer':<15} | {'Porter Stemmer':<15} | {'Lancaster Stemmer':<15}")
print("-" * 73)

for token in cleaned_tokens:
    r_stem = regex_stemmer.stem(token)
    p_stem = porter_stemmer.stem(token)
    l_stem = lancaster_stemmer.stem(token)
    print(f"{token:<20} | {r_stem:<15} | {p_stem:<15} | {l_stem:<15}")
```
