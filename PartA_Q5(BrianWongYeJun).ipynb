This notebook presents an independent implementation of an alternative tokenization technique using the Gensim library. The obtained results are compared with the tokenization approaches implemented in Q1, namely the split() function, Regular Expression, and NLTK word_tokenize().

## Environment Setup

The Gensim library is imported to perform tokenization. Gensim provides a lightweight tokenizer designed for text mining and natural language processing applications.

```python
# Uncomment if Gensim is not installed
# %pip install gensim

from gensim.utils import tokenize

print('Environment setup completed successfully.')
```

## Q5.1 Alternative Tokenization using Gensim

The text corpus from Data_1.txt is loaded and tokenized using Gensim's tokenize() function. Unlike split() and Regular Expression, Gensim automatically extracts valid word tokens while ignoring punctuation.

```python
with open('Data_1.txt', 'r', encoding='utf-8') as file:
    corpus_text = file.read()

gensim_tokens = list(tokenize(corpus_text))

print(f'Gensim Tokenization [Count: {len(gensim_tokens)} tokens]:\n')
print(gensim_tokens)
```

## Q5.2 Compare and Contrast with Q1's Tokenization Methods

To compare the alternative approach with the group's implementation, the same corpus is tokenized using the split() function and NLTK word_tokenize(). The outputs are then compared with Gensim.

```python
import re
from nltk.tokenize import word_tokenize

split_tokens = corpus_text.split()
regex_tokens = re.findall(r"\b\w+\b", corpus_text)
nltk_tokens = word_tokenize(corpus_text)

print('Token Counts')
print('----------------------------')
print(f'Split()           : {len(split_tokens)}')
print(f'Regular Expression: {len(regex_tokens)}')
print(f'NLTK              : {len(nltk_tokens)}')
print(f'Gensim            : {len(gensim_tokens)}')
```

```python
print('Comparison of First 30 Tokens')
print('---------------------------------------')

print('Split():')
print(split_tokens[:30])

print('\nRegular Expression:')
print(regex_tokens[:30])

print('\nNLTK:')
print(nltk_tokens[:30])

print('\nGensim:')
print(gensim_tokens[:30])
```

## Q5.3 Discussion

### Comparison with Q1 Approaches

| Method | Characteristics |
|-------|------------------|
| Split() | Splits text only by whitespace and retains punctuation. |
| Regular Expression | Uses manually defined patterns to extract words. |
| NLTK word_tokenize() | Linguistically aware tokenizer that separates punctuation into individual tokens. |
| Gensim tokenize() | Automatically extracts only meaningful word tokens while ignoring punctuation. |

### Discussion

Gensim's tokenizer provides a lightweight and efficient approach for tokenization. Unlike the split() function, it automatically removes punctuation and extracts only valid word tokens. Compared with Regular Expression, Gensim does not require manually designing tokenization patterns, making the implementation simpler and easier to maintain.

Compared with NLTK word_tokenize(), Gensim focuses on producing clean tokens suitable for text mining, keyword extraction, document clustering and topic modelling. However, it does not preserve punctuation or sentence boundary information, making it less suitable for advanced linguistic analysis such as part-of-speech tagging or syntactic parsing.

Therefore, Gensim is not necessarily better than the methods used in Q1. Instead, it is more suitable for applications where clean word tokens are required, whereas NLTK remains the preferred choice for detailed natural language processing tasks.
